## importando blibliotecas 

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np

import os

print("Todas as bibliotecas foram importadas com sucesso usando seus aliases.")
```

```python
# 'path' é a variável que você já tem com o caminho do Kaggle
# Ex: path = '/home/user/.kaggle/datasets/ayeshasiddiqa123/finance-trends-2020-2025/2b9952a1-5fef-4813-98a7-5e8acede41e5'

# 1. Listamos os arquivos dentro da pasta para encontrar o CSV
try:
    files_in_directory = os.listdir(path)
    csv_file = [f for f in files_in_directory if f.endswith('.csv')][0]

    # 2. Criamos o caminho completo para o arquivo CSV
    full_csv_path = os.path.join(path, csv_file)

    print(f"Arquivo encontrado: {csv_file}")
    print(f"Carregando dados de: {full_csv_path}")

    # 3. Carregamos o arquivo em um DataFrame do Pandas
    df = pd.read_csv(full_csv_path)

    # 4. Verificamos se funcionou, mostrando as 5 primeiras linhas
    print("\nDataFrame carregado com sucesso! As 5 primeiras linhas são:")
    print(df.head())

except (FileNotFoundError, IndexError):
    print("Erro: Nenhum arquivo .csv foi encontrado no caminho especificado.")
    print("Verifique se a variável 'path' está correta e se o dataset foi baixado.")

```
